```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

fig= plt.figure()
ax = fig.add_subplot(111)
ax.patch.set_facecolor('#000000')

E1 = mpatches.Ellipse(xy=(-3,20),width=6,height=2,angle=-45,color='w')
E2 = mpatches.Ellipse(xy=(3,20),width=6,height=2,angle=45,edgecolor='w',facecolor='w')
C1 = mpatches.Circle(xy=(3,0),radius=3,edgecolor='w',facecolor='w')
C2 = mpatches.Circle(xy=(-3,0),radius=3,edgecolor='w',facecolor='w')
C3 = mpatches.Circle(xy=(0,16),radius=4,edgecolor='w',facecolor='w')
C4 = mpatches.Circle(xy=(-1.5,18),radius=0.2,color='k')
C5 = mpatches.Circle(xy=(1.5,18),radius=0.2,color='k')
S1  = mpatches.Rectangle(xy=(-4,0),width=8,height= 16,edgecolor='k',facecolor='w')
S2  = mpatches.Rectangle(xy=(-2,15),width=4.5,height= 1,color='w' )
S3  = mpatches.Rectangle(xy=(-3,-0.1),width=6,height= 0.2,color='w' )
S4 = mpatches.Rectangle(xy=(-3.99,0),width=7.98,height= 3,color='w')
ax.add_patch(E1)
ax.add_patch(E2)
ax.add_patch(C1)
ax.add_patch(C2)
ax.add_patch(C3)
ax.add_patch(C4)
ax.add_patch(C5)
ax.add_patch(S1)
ax.add_patch(S2)
ax.add_patch(S3)
ax.add_patch(S4)
##笑臉
x= np.linspace(-1.5,1.5,50)
y= 0.5*x**2 +14.5
ax.plot(x,y,'-',color='r')
plt.axis('equal')

plt.show()
```

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

plt.xkcd()
fig= plt.figure("Rabbit")
ax = fig.add_subplot(111)
ax.patch.set_facecolor('#000000')

#形狀 Ｅ=橢圓 C=圓形 S=方形
E1 = mpatches.Ellipse(xy=(-3,20),width=6,height=2,angle=-45,color='w')
E2 = mpatches.Ellipse(xy=(3,19.7),width=6,height=2,angle=45,edgecolor='w',facecolor='w')
C1 = mpatches.Circle(xy=(3,0),radius=3,edgecolor='w',facecolor='w')
C2 = mpatches.Circle(xy=(-3,0),radius=3,edgecolor='w',facecolor='w')
C3 = mpatches.Circle(xy=(0,16),radius=4,edgecolor='w',facecolor='w')
C4 = mpatches.Circle(xy=(-1.5,18),radius=0.2,color='k')
C5 = mpatches.Circle(xy=(1.5,18),radius=0.2,color='k')
C6 = mpatches.Circle(xy=(-1.5,18),radius=0.4,edgecolor='k',facecolor='w')
C7 = mpatches.Circle(xy=(1.5,18),radius=0.4,edgecolor='k',facecolor='w')
S1  = mpatches.Rectangle(xy=(-4,0),width=8,height= 16,edgecolor='w',facecolor='w')
#S2  = mpatches.Rectangle(xy=(-2,15),width=4.5,height= ,color='w' )
S2  = mpatches.Rectangle(xy=(-3,-0.1),width=6,height= 0.2,color='w' )
S3 = mpatches.Rectangle(xy=(-3.99,0),width=7.98,height= 3,color='w')


#把形狀塗滿
ax.add_patch(E1)
ax.add_patch(E2)
ax.add_patch(C1)
ax.add_patch(C2)
ax.add_patch(C3)
ax.add_patch(C6)
ax.add_patch(C7)
ax.add_patch(S1)
ax.add_patch(S2)
ax.add_patch(S3)
#ax.add_patch(S4)
ax.add_patch(C4)
ax.add_patch(C5)
##笑臉我這邊使用拋物線
x= np.linspace(-1.5,1.5,50)
y= 0.4*x**2 +15
ax.plot(x,y,'-',color='r')
x2= np.linspace(-4,4,50)
y2= 0.5*x**2 +13
ax.plot(x2,y2,'-',color='r')
x3= np.linspace(-2.5,2.5,50)
y3= 0.18*x**2 +13
ax.plot(x3,y3,'-',color='w')
plt.axis('equal') #把軸變成等長
#寫字
plt.text(-12.5, 9, "Rabb", size=50, rotation=0.,color='w',
         ha="center", va="center",)
plt.text(6.5, 9, "t", size=50, rotation=0.,color='w',
         ha="center", va="center",)
plt.text(10, 8.7, "!", size=50, rotation=0.,color='w',
         ha="center", va="center",)
plt.savefig("hw01.png",dpi=120)#輸出圖片存擋
plt.show()#輸出圖片


```
